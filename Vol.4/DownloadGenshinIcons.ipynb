# Crawl fandom website

## Front

```python
file_path_export = '/content/drive/MyDrive/Genshin/Character Surveys/Vol. 4/Icons/Front/'
```

```python
url_to_crawl = f'https://genshin-impact.fandom.com/wiki/Genshin_Impact_Wiki?file=REPLACEME_Icon.png'
```

```python
sample_image_url = 'https://static.wikia.nocookie.net/gensin-impact/images/f/f6/Yumemizuki_Mizuki_Icon.png'
```

```python
'https://static.wikia.nocookie.net/gensin-impact/images/f/f6/Yumemizuki_Mizuki_Icon.png/revision/latest/scale-to-width-down/74?cb=20250212014631'
```

```python
# prompt: # I need to crawl the Genshin Impact fandom website and batch download a bunch of images. For each character in list_all_genshin_chars, generate a website URL by replacing 'REPLACEME' in url_to_crawl with the character name. Then for each website URL, crawl the source code for that URL and look for an image URL with the form 'https://static.wikia.nocookie.net/gensin-impact/images/SOMETHING/CHARACTERNAME_Icon.png', where the SOMETHING can vary for each character, and CHARACTERNAME is the corresponding string in list_all_genshin_chars. Download that image and save it to the file_path_export path.

import requests
import re
import os

for char_name in list_all_genshin_chars:
  # Construct the URL for the character's icon page
  icon_url = url_to_crawl.replace('REPLACEME', char_name)

  try:
    # Fetch the content of the icon page
    response = requests.get(icon_url)
    response.raise_for_status() # Raise an HTTPError for bad responses (4xx or 5xx)
    html_content = response.text

    # Look for the image URL pattern in the HTML content
    # The pattern looks for URLs starting with 'https://static.wikia.nocookie.net/gensin-impact/images/'
    # followed by any characters (the SOMETHING part), then the character name
    # and '_Icon.png'. It captures the full image URL.
    image_url_match = re.search(rf'https://static\.wikia\.nocookie\.net/gensin-impact/images/[^"]*/{re.escape(char_name)}_Icon\.png', html_content)

    if image_url_match:
      image_url = image_url_match.group(0)
      print(f"Found image URL for {char_name}: {image_url}")

      # Download the image
      image_response = requests.get(image_url)
      image_response.raise_for_status()

      # Determine the file name and path
      # Extract the file name from the image URL
      file_name = os.path.basename(image_url)
      file_path = os.path.join(file_path_export, file_name)

      # Save the image to the specified path
      with open(file_path, 'wb') as f:
        f.write(image_response.content)
      print(f"Downloaded and saved {file_name} to {file_path}")

    else:
      print(f"Could not find image URL for {char_name} on page: {icon_url}")

  except requests.exceptions.RequestException as e:
    print(f"Error fetching or processing {char_name}'s icon page ({icon_url}): {e}")
  except Exception as e:
    print(f"An unexpected error occurred for {char_name}: {e}")
```

```python
# prompt: # I need to crawl the Genshin Impact fandom website and batch download a bunch of images. For each character in list_all_genshin_chars, generate a website URL by replacing 'REPLACEME' in url_to_crawl_side with the character name. Then for each website URL, crawl the source code for that URL and look for an image URL with the form 'https://static.wikia.nocookie.net/gensin-impact/images/SOMETHING/CHARACTERNAME_Side_Icon.png', where the SOMETHING can vary for each character, and CHARACTERNAME is the corresponding string in list_all_genshin_chars. Download that image and save it to the file_path_export_side path.

import pandas as pd
from google.colab import drive
import requests
import re
import os
# Assuming drive is already mounted and list_all_genshin_chars is populated from the preceding code

# Create the output directory if it doesn't exist
os.makedirs(file_path_export_side, exist_ok=True)

for char_name in list_all_genshin_chars:
  # Construct the URL for the character's gallery page, which contains the side icon
  # Note: The base URL structure for the side icon is slightly different, often within a Gallery page
  # Let's adjust the URL construction to reflect this common structure if it varies
  # Based on the example 'https://genshin-impact.fandom.com/wiki/Chiori/Gallery?file=REPLACEME_Side_Icon.png',
  # it seems we still navigate to the character's Gallery page.
  gallery_url = f'https://genshin-impact.fandom.com/wiki/{char_name}/Gallery'

  try:
    # Fetch the content of the gallery page
    response = requests.get(gallery_url)
    response.raise_for_status() # Raise an HTTPError for bad responses (4xx or 5xx)
    html_content = response.text

    # Look for the image URL pattern in the HTML content
    # The pattern looks for URLs starting with 'https://static.wikia.nocookie.net/gensin-impact/images/'
    # followed by any characters (the SOMETHING part), then the character name
    # and '_Side_Icon.png'. It captures the full image URL.
    image_url_match = re.search(rf'https://static\.wikia\.nocookie\.net/gensin-impact/images/[^"]*/{re.escape(char_name)}_Side_Icon\.png', html_content)

    if image_url_match:
      image_url = image_url_match.group(0)
      print(f"Found side image URL for {char_name}: {image_url}")

      # Download the image
      image_response = requests.get(image_url)
      image_response.raise_for_status()

      # Determine the file name and path
      # Extract the file name from the image URL
      file_name = os.path.basename(image_url)
      file_path = os.path.join(file_path_export_side, file_name)

      # Save the image to the specified path
      with open(file_path, 'wb') as f:
        f.write(image_response.content)
      print(f"Downloaded and saved {file_name} to {file_path}")

    else:
      print(f"Could not find side image URL for {char_name} on page: {gallery_url}")

  except requests.exceptions.RequestException as e:
    print(f"Error fetching or processing {char_name}'s gallery page ({gallery_url}): {e}")
  except Exception as e:
    print(f"An unexpected error occurred for {char_name}: {e}")
```

## Small images

```python
file_path_export_small = '/content/drive/MyDrive/Genshin/Character Surveys/Vol. 4/Icons/Small/'
```

```python
url_to_crawl_small = 'https://sunderarmor.com/GENSHIN/Characters/1/REPLACEME.png' # %20
```

```python
dict_character_name_swap = {
    'Tartaglia' : 'Childe',
    'Kaedehara_Kazuha' : 'Kazuha',
    'Kamisato_Ayaka' : 'Ayaka',
    'Raiden_Shogun' : 'Raiden',
    'Sangonomiya_Kokomi' : 'Kokomi',
    'Arataki_Itto' : 'Itto',
    'Kamisato_Ayato' : 'Ayato',
    'Kujou_Sara' : 'Sara',
    'Shikanoin_Heizou' : 'Heizou',
}
```

```python
# prompt: # For each character in the list list_all_genshin_chars, create a URL by replacing 'REPLACEME' in the url_to_crawl_small with the corresponding character entry, and also replace any underscore in the character name with %20. Download the PNG found at that url and save it to file_path_export_small.

# Create the output directory if it doesn't exist
os.makedirs(file_path_export_small, exist_ok=True)

for char_name in list_all_genshin_chars:
  # Replace underscores with %20 in the character name for the URL
  if char_name in dict_character_name_swap:
    char_name = dict_character_name_swap[char_name]
  char_name_formatted = char_name.replace('_', '%20')

  # Construct the URL for the small image
  image_url = url_to_crawl_small.replace('REPLACEME', char_name_formatted)

  try:
    # Download the image
    image_response = requests.get(image_url, stream=True)
    image_response.raise_for_status() # Raise an HTTPError for bad responses (4xx or 5xx)

    # Determine the file name and path
    # Use the original character name for the file name, replacing underscores with spaces
    file_name_base = char_name.replace('_', ' ')
    file_name = f"{file_name_base}.png" # Assuming the downloaded file is a PNG
    file_path = os.path.join(file_path_export_small, file_name)

    # Save the image to the specified path
    with open(file_path, 'wb') as f:
      # Use stream to handle potentially large files
      for chunk in image_response.iter_content(chunk_size=8192):
        f.write(chunk)

    print(f"Downloaded and saved {file_name} to {file_path}")

  except requests.exceptions.RequestException as e:
    print(f"Error downloading image for {char_name} from {image_url}: {e}")
  except Exception as e:
    print(f"An unexpected error occurred for {char_name}: {e}")
```
